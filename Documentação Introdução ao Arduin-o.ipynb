## 01 - Criar um código para mudar a velocidade do led



* Declaração de Variável:

* int led = 13;: Declara uma variável inteira chamada led e a inicializa com o valor 13, indicando que o LED está conectado à porta digital 13.

* Função setup():

* pinMode(led, OUTPUT);: Configura a porta do LED como uma saída. Isso é necessário para que o Arduino possa enviar um sinal para acender o LED.

* Função loop():

* digitalWrite(led, HIGH);: O programa envia um sinal de tensão alta para a porta do LED, fazendo com que ele acenda.

* delay(10000);: O programa pausa por 10000 milissegundos, o que equivale a 10 segundos. Durante esse tempo, o LED permanece aceso.

* digitalWrite(led, LOW);: O programa envia um sinal de tensão baixa para a porta do LED, fazendo com que ele apague.

* delay(10000);: O programa pausa novamente por 10 segundos. Durante esse tempo, o LED permanece apagado.

## 02 - Criar um padrão (Fazendo o LED piscar 3 vezes rápido e depois 1 vez devagar).




* Declaração de Variável:

* int led = 13;: Declara uma variável inteira led e a inicializa com o valor 13, que é a porta digital onde o LED será conectado.

* Função setup():

* pinMode(led, OUTPUT);: Configura a porta digital do led como uma saída. Isso permite que o microcontrolador envie sinais elétricos para o LED, controlando se ele está ligado ou desligado.

* Função loop():

* A função loop() é executada continuamente, indefinidamente, depois que a função setup() é concluída. O código fará com que o LED pisque no padrão definido repetidamente.

* for (i = 0; i < 3; i++): O programa executa um loop três vezes.

* digitalWrite(led, HIGH);: Acende o LED.

* delay(200);: O programa aguarda 200 milissegundos.

* digitalWrite(led, LOW);: Apaga o LED.

* delay(200);: O programa aguarda mais 200 milissegundos.

* delay(900);: Após as três piscadas rápidas, há uma pausa de 900 milissegundos.

* digitalWrite(led, HIGH);: O LED acende novamente.

* delay(1500);: O programa aguarda 1500 milissegundos, ou 1.5 segundos, mantendo o LED aceso.

* digitalWrite(led, LOW);: Apaga o LED.

* delay(1500);: O programa aguarda mais 1.5 segundos antes de recomeçar o loop.

## 03 - Adicionar um botão O LED só deve piscar se um botão estiver pressionado com uma variável controlando o tempo de acendimento do LED



Declaração de Variáveis:

* int led = 13;: Declara uma variável inteira chamada led e a inicializa com o valor 13. A porta digital 13 é comumente usada para LEDs de saída.

* int botao = 12;: Declara uma variável inteira chamada botao e a inicializa com o valor 12, indicando que o botão de controle está conectado à porta digital 12.

* int tempo = 1000;: Declara uma variável inteira chamada tempo e a inicializa com 1000 milissegundos (ou 1 segundo). Este valor é usado como a base para os tempos de delay.

* Função setup():

* pinMode(led, OUTPUT);: Configura a porta do LED como uma saída, permitindo que o microcontrolador a use para enviar energia e acender o LED.

* pinMode(botao, INPUT_PULLUP);: Configura a porta do botão como uma entrada com um resistor de pull-up interno ativado. Isso significa que a porta digital fica em um estado "alto" (HIGH) por padrão. Quando o botão é pressionado e a porta é conectada ao terra, o estado da porta muda para "baixo" (LOW).

* Função loop():

* if (digitalRead(botao) == LOW): Esta é a condição principal. O programa verifica se o botão foi pressionado. Como a porta foi configurada com pull-up, ela retorna LOW quando o botão é pressionado.

* for (i = 0; i < 3; i++): Se o botão for pressionado, um loop for é executado três vezes.

* digitalWrite(led, HIGH);: Acende o LED.

* delay(tempo / 5);: Aguarda um tempo de 200 milissegundos (1000 / 5).

* digitalWrite(led, LOW);: Apaga o LED.

* delay(tempo / 5);: Aguarda mais 200 milissegundos.

* delay(tempo - 100);: Após o loop for terminar, o programa aguarda um tempo de 900 milissegundos (1000 - 100) antes de executar a próxima ação. Este atraso cria uma pausa entre a sequência rápida de piscadas e a última piscada lenta.

* digitalWrite(led, HIGH);: Acende o LED pela última vez.

* delay(tempo + 500);: Aguarda um tempo de 1500 milissegundos (1000 + 500), mantendo o LED aceso por mais tempo.

* digitalWrite(led, LOW);: Apaga o LED.

* delay(tempo + 500);: Aguarda mais 1500 milissegundos
