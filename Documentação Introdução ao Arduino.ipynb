## Criar um código para mudar a velocidade do led



começamos criando a variável led sendo ligada na porta 13, usamos o delay para cronometrar a velocidade e o DigitalWrite para identificar a variável led e ligar ela com high e desligar com low   




----
## Criar um padrão (Fazendo o LED piscar 3 vezes rápido e depois 1 vez devagar).




declaramos a variável para o LED na porta 13, colocamos um pinMode de saída para o LED, criando um for com um liga e desliga com ambos com delay de 2 milisegundos sendo o piscar 3 vezes, fora deste for foi criado um outro delay de 9ms e um liga e desliga com delay de 1s e meio sendo a 1 vez devagar
